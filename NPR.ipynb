```python
#Bibliotecas utilizadas:
import numpy as np 
import matplotlib.pyplot as plt

cont=-1
NPR_Perc=[]
Acum=[]
legenda=[]

#Entrada de dados:
input_string = input('Entre com os dados separados por espaço:')
print("\n")
NPR= input_string.split()#Transforma os dados de entrada em uma lista de strings utilizando como diferenciador o espaço entre eles.

#Tranformação de string em float:
for i in range(len(NPR)): #Corre todos os elementos da lista de strings criada (NPR).
    NPR[i] = float(NPR[i])#Transforma cada elemento para o tipo float.
    legenda.append(str(i))#Para cada elemento da lista NPR recebe um elemento i tipo string.

NPR.sort(reverse=True) # Inverte a lista NPR.

#Valores percentuais das NPR:
for j in NPR: #Corre todos os elementos da lista NPR.
  k=(j/sum(NPR)*100) #Calcula o valor percentual de cada elemento.
  NPR_Perc.append(k) #Armazena os dados percentuais em uma nova lista.

#Curva de NPR percentual acumulado:
for l in NPR_Perc: #Corre todos os valores em NPR_Perc.
  cont=cont+1 #Contador.
  if cont==0: #Condição para receber o primeiro elemento.
    Acum.append(l)#Primeiro elemento de NPR_Perc adicionado a lista Acum.
  else: #Para todos os casos que não se enquadram na primeira condição.
    Acum.append(l+Acum[cont-1]) #Adiciona a lista Acum a soma de um elemento de NPR_Perc com o último elemento de Acum.

fig, ax = plt.subplots() #Funções para plotagem.
plt.rcParams["figure.figsize"] = (10, 5) #Para controlar as dimensões da figura.
#Observação: Uma vez definidos os parâmetros da linha acima, rode o programa duas vezes.

#Gerando o gráfico de barras
ax.bar(legenda, NPR_Perc, color="C0", width=0.5, edgecolor='black') #Parâmetros.
#Observação: Apenas os dois primeiros são necessários. Os outros se referem a cor, largura e cor das bordas das barras respectivamente.

#Plotando a curva de NPR acumulado:
plt.plot(legenda, Acum, color = "red", marker = "D", label='NPR acumulado') #Parâmetros.
plt.legend() #Necessário para que a legenda escolhida apareça na figura.
#Observação: Apenas os dois primeiros são necessários. Os outros se referem a cor, marcador e legenda respectivamente.

ax.grid() #Adiciona grades ao fundo da figura.
ax.tick_params(axis="y", colors="black", labelsize=12) #Controla o tamanho, cor e fonte dos elementos do eixo y.
ax.set_ylabel('NPR[%]', size=12,fontweight='bold') #Controla a legenda do eixo y.


plt.show()


```
