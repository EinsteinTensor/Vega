```python
# Bibliotecas e funções importadas
import numpy as np
import matplotlib.pyplot as plt
import functools

################################################################################

#Criação de listas, arquivos e outros objetos relevantes para este código.
lista1=[]; lista2=[]; lista3=[]; lista4=[]; A=[]; B=[]; Cor=[]; legenda=[]; colorir=[]
dic1={}
contador=0
cont=0
contagem=[]

Arq1=open('Vega1.txt', 'w')
Arq1.close()
Arq2=open('Vega2.txt', 'w')
Arq2.close()
################################################################################

#Corpo principal

controlador='sim'
while controlador!='não':
  contador+=1
  contagem.append(1)

  # Para receber os dados
  input_string = input('Entre com os dados do intervalo de dose separados por espaço:')
  print("\n")
  dose = input_string.split()

  for i in range(len(dose)):
    #convert each item to float type
    dose[i] = float(dose[i])

  input_string=input('Entre com os dados da frequência total predita separados por espaço:')
  print("\n")
  frequência=input_string.split()

  cor=input('Qual cor deseja nessa curva?')
  Cor.append(str(cor))

  pais=input('Insira a legenda:')
  legenda.append(str(pais))

  frequência.reverse()
  for i in range(len(frequência)):
    #convert each item to float type
    frequência[i] = float(frequência[i])

  for indice in range(0, len(dose)):
    #Bloco 1
    if indice<len(dose)-1:
      Dose=np.linspace(dose[indice], dose[indice+1], num=5)
      for elemento in Dose:
        lista1.append(elemento)
        lista2.append(frequência[indice])

      Arq1= open("Vega1.txt", 'a')
      for termo in lista1:
        Arq1.write(str(termo))
        Arq1.write("\n")
      Arq1.close()

      Arq2= open("Vega2.txt", 'a')
      for termo in lista2:
        Arq2.write(str(termo))
        Arq2.write("\n")
      Arq2.close()

    if indice==len(dose)-1:
      Dose=np.linspace(dose[indice], 1*pow(10, 5), num=5) #Caso queira modificar o limite superior, basta modificar 1*pow(10,5) por aquilo que deseja
      for elemento in Dose:
        lista1.append(elemento)
        lista2.append(frequência[indice])

      Arq1= open("Vega1.txt", 'a')
      for termo in lista1:
        Arq1.write(str(termo))
        Arq1.write("\n")
      Arq1.close()

      Arq2= open("Vega2.txt", 'a')
      for termo in lista2:
        Arq2.write(str(termo))
        Arq2.write("\n")
      Arq2.close()

    #bloco 2
    if indice<len(dose)-1:
      lista3=np.linspace(frequência[indice], frequência[indice+1], num=5)
      Arq2= open("Vega2.txt", 'a')
      for termo in lista3:
        Arq2.write(str(termo))
        Arq2.write("\n")
      Arq2.close()

      for j in range(0, len(lista1)):
        lista4.append(lista1[-1])
      Arq1= open("Vega1.txt", 'a')
      for termo in lista4:
        Arq1.write(str(termo))
        Arq1.write("\n")
      Arq1.close()

    if indice==len(dose):
      lista3=np.linspace(frequência[indice], 1*pow(10, 5), num=5) #Caso queira modificar o limite superior, basta modificar 1*pow(10,5) por aquilo que deseja
      Arq2= open("Vega2.txt", 'a')
      for termo in lista3:
        Arq2.write(str(termo))
        Arq2.write("\n")
      Arq2.close()

      for j in range(0, len(lista1)):
        lista4.append(lista1[-1])
      Arq1= open("Vega1.txt", 'a')
      for termo in lista4:
        Arq1.write(str(termo))
        Arq1.write("\n")
      Arq1.close()

    lista1=[]; lista2=[]; lista3=[]; lista4=[]

  Arq1= open("Vega1.txt")
  linhas = Arq1.readlines()
  for linha in linhas:
      A.append(float(linha))
  Arq1.close()

  Arq2= open("Vega2.txt")
  linhas = Arq2.readlines()
  for linha in linhas:
      B.append(float(linha))
  Arq2.close()

  Arq1= open("Vega1.txt", "r+")
  Arq1.truncate(0)
  Arq1.close()

  Arq2= open("Vega2.txt", "r+")
  Arq2.truncate(0)
  Arq2.close()

  dic1['A'+str(contador)]=A
  dic1['B'+str(contador)]=B

  A=[]; B=[]

  controlador=input('Digite sim caso queira continuar:') # Repete todo o processo de aquisição caso seja digitado qualquer coisa diferente de 'não'.

fig, ax=plt.subplots()

# Para plotar as curvas
for k in range(1, len(contagem)+1):
  X=dic1.get('A'+str(k))
  Y=dic1.get('B'+str(k))
  Y.reverse()
  plt.plot(X, Y ,color=Cor[k-1], label=legenda[k-1])

  #Para eliminar legendas repetidas
  handles, labels = plt.gca().get_legend_handles_labels()
  by_label = dict(zip(labels, handles))
  plt.legend(by_label.values(), by_label.keys())

# Para colorir a área entre curvas de um mesmo conjunto de dados
for l in range(1, len(contagem)+1):
    T1=dic1.get('A'+str(l))
    V1=dic1.get('B'+str(l))
    for m in range(l+1, len(contagem)+1):
      T2=dic1.get('A'+str(m))
      V2=dic1.get('B'+str(m))
      if functools.reduce(lambda i, j : i and j, map(lambda m, k: m == k, T1, T2), True) : # Se T1 for igual a T2 (Curvas tipo Grã-Bretanha)
        ax.fill_between(T1, V1, V2, color=Cor[m-1], alpha=0.2)
      else: # Caso T1 não seja igual a T2 (Curvas tipo ICRP 64)
        if legenda[l-1]==legenda[m-1]:
          ax.fill_between(T1, V1, V2, color=Cor[m-1], alpha=0.2)
          ax.fill_betweenx(V2, T1, T2, color=Cor[m-1], alpha=0.2)
          


# Questões do gráfico     
plt.xscale('log') # Eixo x em escala logaritmica
plt.yscale('log') # Eixo y em escala logaritmica
ax.grid()# Grades no fundo do gráfico
ax.set_facecolor('whitesmoke') # Colorir fundo do gráfico
plt.xlabel('Dose máxima efetiva (mSv)') # Legenda eixo x
plt.ylabel('Frequência total predita (por ano)') # Legenda eixo y
```
